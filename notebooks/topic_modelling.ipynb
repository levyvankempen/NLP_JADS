```python
import pandas as pd
from gensim import corpora
from gensim.models.ldamodel import LdaModel
import logging
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# Load the preprocessed dataframe
df = pd.read_csv('../data/preprocessed_article_data.csv')
preprocessed_articles_text = df['content']
```

```python
texts = preprocessed_articles_text.tolist()
```

```python
texts = [doc.split() for doc in texts]
```

```python
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
num_topics = 10  # You can adjust this based on your requirement
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
```

```python
topics = lda_model.print_topics(num_words=10)
for topic in topics:
    print(topic)

```

```python
lda_model.save('../data/init_lda_model')
```
